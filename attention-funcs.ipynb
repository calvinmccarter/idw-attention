```python
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
plt.figure(figsize=(4,3.2));

n = 1000;
x = np.linspace(0, 5, n);
X = np.ones((n, 2));
X[:, 0] = x;

Y = torch.softmax(torch.tensor(-(X**2)), axis=1).numpy();
plt.plot(x, Y[:, 0], label='Neg Dist ($p=2$)');

Y = torch.softmax(torch.tensor(np.exp(-0.5*X)), axis=1).numpy();
plt.plot(x, Y[:, 0], label='Gaussian Dist ($p=2$)');

Y = torch.softmax(torch.tensor(1/(1e-3 + X)), axis=1).numpy();
plt.plot(x, Y[:, 0], label='Inv Dist ($p=2, ϵ=10^{-3}$)');

Y = torch.softmax(torch.tensor(-np.log(1e-3 + X)), axis=1).numpy();
plt.plot(x, Y[:, 0], label='IDW ($p=2, ϵ=10^{-3}$)');

plt.grid(True);
plt.yticks([0, 0.5, 1.0]);
plt.ylabel('weight');
plt.xlabel('distance');
plt.legend();
plt.savefig('attention-dist-func.pdf');
```
